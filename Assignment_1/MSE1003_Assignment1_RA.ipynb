```python
from mp_api.client import MPRester
import pandas as pd

API_KEY = "YOUR_API_KEY_HERE"

data = []

with MPRester(API_KEY) as mpr:

    # Conductors (band gap = 0)
    conductors = mpr.materials.summary.search(
        band_gap=(0, 0),
        fields=["material_id", "formula_pretty", "band_gap"],
        num_chunks=20
    )[:2000]

    # Semiconductors (0 < band gap ≤ 3)
    semiconductors = mpr.materials.summary.search(
        band_gap=(0.01, 3),
        fields=["material_id", "formula_pretty", "band_gap"],
        num_chunks=20
    )[:2000]

    # Insulators (band gap > 3)
    insulators = mpr.materials.summary.search(
        band_gap=(3.01, 10),
        fields=["material_id", "formula_pretty", "band_gap"],
        num_chunks=20
    )[:2000]

    for m in conductors:
        data.append([m.material_id, m.formula_pretty, m.band_gap, "conductor"])

    for m in semiconductors:
        data.append([m.material_id, m.formula_pretty, m.band_gap, "semiconductor"])

    for m in insulators:
        data.append([m.material_id, m.formula_pretty, m.band_gap, "insulator"])

df = pd.DataFrame(
    data,
    columns=["material_id", "composition", "band_gap_eV", "class"]
)

print(df.head())
print(len(df))
```
